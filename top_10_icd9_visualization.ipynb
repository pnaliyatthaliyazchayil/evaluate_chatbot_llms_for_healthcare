```python
import pandas as pd
import numpy as np
from sklearn.metrics import multilabel_confusion_matrix
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('C:/Users/Parvati/Desktop/mimic/llama_final_with_diagnosis_readmisison_sick_comparison.csv')
# Strip trailing spaces from the icd9_codes column, split by commas, and explode
all_icd_codes = df['converted_icd_code'].str.strip().str.split(',').explode().str.strip()

# Regex to add the period to get the ICD9 codes
all_icd_codes = all_icd_codes.replace(r'^(\d{3})\.*(\d.*)?$', r'\1.', regex=True)

#Combine counts for codes with the same first 3 digits
all_icd_codes_3digit = all_icd_codes.str[:3]  

# Get the count of each unique ICD-9 code by rollup to parent using first 3 digits
icd_code_counts = all_icd_codes_3digit.value_counts()

# Print the result
print(icd_code_counts)

# Plotting the top 10 most common ICD codes
top_10_icd_codes = icd_code_counts.head(10)

# Create a horizontal bar graph (sorted in descending order)
plt.figure(figsize=(10, 6))
top_10_icd_codes.sort_values(ascending=True).plot(kind='barh', color='skyblue')

# Add labels and title
plt.xlabel('Counts of the ICD-9 Codes', fontsize=14)
plt.ylabel('ICD-9 Code Label', fontsize=14)
plt.title('Top 10 Most Common ICD-9 Codes for MIMIC-IV sample', fontsize=16)

# Show the plot
plt.show()
```
